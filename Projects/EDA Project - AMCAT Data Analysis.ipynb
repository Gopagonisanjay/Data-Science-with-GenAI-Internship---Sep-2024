## 2. DATA CLEANING

### Type Casting

* We can see the DOJ and DOL are given in timestamp format. As per our objective,we only need date,we will convert timestamp into date using datetime module.

* In DOL column,We can see the value 'present' . We will convert this into the present date for our analysis

```python
df['DOL'].replace('present',dt.datetime.today(),inplace=True)
df['DOL'] = pd.to_datetime(df['DOL']).dt.date
```

```python
df['DOJ'] = pd.to_datetime(df['DOJ'])
df['DOL'] = pd.to_datetime(df['DOL'])

df["period"] = (df['DOL'] - df['DOJ']).dt.days
```

```python
df.head()
```

###### Research Questions
Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
# Categorical Vs Catogorical using Crosstab
Crosstab = pd.crosstab(columns=df['Specialization'], index=df['Gender'])
Crosstab
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2, p,dof,ex = chi2_contingency(Crosstab)
```

```python
print(f'Chi-square Statistic: {chi2}')
print(f'p-value: {p}')
```

```python
alpha = 0.05
if p < alpha:
    print('There is a significant relationship between gender and specialization.')
else:
    print('There is no significant relationship between gender and specialization.')
```
